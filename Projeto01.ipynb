```python
#Mudando o nome das columas
df = df.rename(columns={"brand":"Marca", "km_driven":"Quilometros_dirigidos", "fuel":"Combustivel", "owner":"Dono", "selling_price":"Preco_venda"})
```

```python
#Visualizando os valores nas colunas
df["Marca"].unique()
```

```python
#Contagem de quantas marcas usam cada combustivel
df.groupby("Combustivel")["Marca"].nunique()
```

```python
#Informações Estastisticas
df.describe()
```

```python
#Verifica a quilometragem media em cada combustivel
df.groupby("Combustivel")["Quilometros_dirigidos"].mean()
```

```python
#Verifica a quilometragem total de cada marca de carro
df.groupby("Marca")["Quilometros_dirigidos"].sum()
```

```python
#Verifica a quilometragem media de cada marca de carro
df.groupby("Marca")["Quilometros_dirigidos"].mean()
```

```python
#Conta a quantidade de vezes que X marca aparece
df["Marca"].value_counts(ascending=False)
```

```python
#Quantidade de vezes que Marca aparecene na tabela em barra horizontal
df["Marca"].value_counts().plot.barh();
```

```python
#Verifica a quilometragem media de cada marca de carro em barra
df.groupby("Marca")["Quilometros_dirigidos"].mean().plot.barh();
```

```python
#Alterando o estilo
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
plt.style.use("ggplot")
```

```python
#Adicionando um titulo ao grafico, mudando o nome dos eixos e a cor da barras
#https://matplotlib.org/stable/gallery/color/named_colors.html
df["Marca"].value_counts().plot.barh(title="Quantidade de carros por Marca", color="peru")
plt.ylabel("Marca")
plt.xlabel("Total_Marcas");
```

```python
#Seleciona apenas as linhas que possui a Marca Maruti
#Verificando o total dos tipos de Combustivel dos carros da Marca Maruti
df_Maruti =  df[df["Marca"] == "Maruti"]
df_Maruti["Combustivel"].value_counts().plot.barh(title="Quantidade de cada tipo de Combustivel");
plt.ylabel("Combustivel")
plt.xlabel("Quantidade")
```

```python
#Verificando a media do quilometros dirigidos dos combustiveis dos carros da Marca Maruti
#Salva o grafico em png
df_Maruti.groupby(df_Maruti["Combustivel"])["Quilometros_dirigidos"].mean().plot.bar(title="Media""\n""Combustivel x Carros Maruti");
plt.savefig("Grafico Media Combustiveis x Carros Marca Maruti")
```
